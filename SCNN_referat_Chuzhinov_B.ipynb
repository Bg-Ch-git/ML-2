***Сферические сверточные нейронные сети***

**Часть 1. Общие вопросы**

1.1 Какую задачу решают авторы?

В статье рассматривается проблема анализа сферических изображений, т.е. изображений пространства, полученных в результате его проекции на сферу. В качестве подхода к решению рассматриваемой автормами проблемы в статье предлагается модифицировать сверточные нейронные сети. Основная проблема, возникающая при рассмотрении сферических изображений, это представление этого изображения в виде начальных данных. В силу того, что сфера не изометрична плоскости, невозможно представить поворот сферы как сдвиг плоскости на какой-то вектор без существенных искажений. А научиться хорошо представлять вращения сферы необходимо, т.к. очевидно, что на один и тот же объект можно смотреть с "разных ракурсов". Вторая цель авторов $-$ корректно определить свертку для полученных представлений сферических изображений.

1.2 Какова основная идея предлагаемого авторами решения поставленной задачи?

Двумерную плоскую свертку можно рассматривать как функцию над $\mathbb{Z}^2$, которая сопоставляет каждому вектору (сдвигу фильтра) число (степень схожести с фильтром):

$$[im*f](x)=⟨im, L_x(f)⟩,$$ где $x\in \mathbb{Z}^2$ - вектор сдвига, $im$ - изображение, $f$ - фильтр, а $L_x$ - оператор сдвига на вектор $x$ и угловые скобки - это скалярное произведение.

Аналогично хочется поступить и со сферическими изображениями. Но есть проблема. Группа $\mathbb{Z}^2$, действующая на дискретной плоскости, изоморфна этой самой плоскости, что нельзя сказать о $SO(3)$ и $S^2$. Из-за этого приходится думать о двух видах сверток: сферических (Spherical Correlation) и вращательных (Rotation Group Correlation). Оба вида сверток должны быть функциями над группой изометрий сферы, то есть над $SO(3)$. При этом нужно определить операторы переноса и для сферы и для группы. Если для сферы все понятно - оператором переноса является естественное действие группы $SO(3)$ на сфере, то для группы оператор переноса определяется менее интуитивно:

$$L_R(f)(Q)=f(R^{-1}Q),$$

где $R,Q\in SO(3)$ - повороты, $f\colon SO(3)\to \mathbb{R}^k$ - "поворотный" фильтр.

После построения строгого математического аппарата, авторы дискретизируют сферу и группу $SO(3)$, используя сетку на сферических координатах и углах Эйлера соответственно. Для эффективных вычислений авторы применяют быстрое преобразование Фурье, модифицированное для сферического случая.

1.3 Каковы результаты, полученные авторами?

**Часть 2. Специальные вопросы**

1. В чём смысл понятия эквивариантности?

Как я понял, операторы переноса должны быть вполне естественными. Должно выполняться следующее свойство. Свертка фильтра с перенесенной (повернутой) картинкой должна совпадать с перенесенной сверткой фильтра с исходной картинкой. То есть с одной стороны у нас есть картинка и мы ее повернули, а потом свернули с фильтром, например, описывающим кота, и получили места, где потенциально есть коты. А с другой стороны, мы можем искать котов на исходной картинке. И результат должен остаться таким же, то есть места расположения котов должны наложиться друг на друга при повороте.

2. Зачем нужны и чем отличаются __Spherical Correlation__ и __Rotation Group Correlation__?

Вроде это просто из-за неизоморфности $SO(3)$ и $S^2$. То есть на входе у вас сферическая картинка, которая является отображением из сферы в $\mathbb{R}^k$, но после первой свертки вы получаете функцию уже на $SO(3)$. Поэтому нужно две разных свертки.

3. Если мы хотим решать задачу классификации или регрессии, нам нужно, чтобы голова сети предсказывала числа или логиты. Как это сделать, если внутри сети возникают всякие сферические свёртки?

Результат свертки - это функция над $SO(3)$, но для решения задач мы дискретизируем $SO(3)$, получая некоторые матрицы и дальше спокойно с ними работаем. Если нужно брать максимум по $SO(3)$, как, например, в max-pooling мы берем дискретный максимум. Дискретизация, как ранее писалось, осуществляется с помощью введения сетки на углах Эйлера.
